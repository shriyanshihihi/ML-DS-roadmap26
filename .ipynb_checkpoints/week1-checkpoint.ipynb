```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("seaborn-v0_8")
```

# Linear regression 

```python
np.random.seed(42) 
# This sets a fixed seed so that the random numbers generated are the same every time you run the code.
#42 has no significance 

X = np.random.rand(100,1)*10
# np.random.rand(100, 1) generates 100 random values between 0 and 1

y = 3*X + 4 + np.random.randn(100,1)*2
# np.random.randn(100, 1) generates Gaussian noise (mean = 0, std = 1)
# Multiplying by 2 adds noise with standard deviation = 2
# Makes the data more realistic
```

```python
plt.scatter(X,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("synthetic linear data")
plt.show()
```

```python
def linear_regression_gd(X,y,lr=0.01,epochs=1000):
    n=len(y)
    w=0
    b=0
    losses=[]
    for i in range(epochs):
        y_pred=w*X+b
        error=y_pred-y
        dw=(1/n)*np.sum(error*X)
        db=(1/n)*np.sum(error)
        w -=lr*dw  #updated weight and bias 
        b -=lr*db
        loss=(1/n)*np.sum(error**2)
        losses.append(loss)
    return w,b,losses
```

```python
w,b,losses = linear_regression_gd(X,y)
print("learned weight:", w)
print("learned bias:" ,b)
```

```python
plt.plot(losses)
plt.xlabel("epoch")
plt.ylabel("mse loss")
plt.title("training loss curve")
plt.show()
```
